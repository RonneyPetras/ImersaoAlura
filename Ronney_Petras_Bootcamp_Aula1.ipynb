```
import pandas as pd
pd.options.display.float_format = "{:.2f}".format
```

##Abrindo os dados com os parametros necessários

```
df = pd.read_csv("https://github.com/alura-cursos/agendamento-hospitalar/blob/main/dados/A151346189_28_143_208.csv?raw=true", 
                 encoding="iso-8859-1", 
                 sep=";",
                 skiprows=3,
                 skipfooter=12,
                 engine="python",
                 thousands=".",
                 decimal=",")
df.head()
```

##Checando os tipos dos dados

```
df.dtypes
```

##Alguns dados estão sem valores, precisamos tratar os valores vazios

```
df[['Unidade da Federação','2020/Ago']].head()
```

```
df = df.replace("-",0.0)
df['2020/Ago'] = df['2020/Ago'].astype("float64")
```

```
df['2008/Ago'].mean()
```

```
df.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,4))
```

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
df = df.sort_values('2008/Ago')
ax = df.plot(x="Unidade da Federação", y="2008/Ago", 
             kind="bar", 
             figsize=(10,4))

tf = ticker.StrMethodFormatter("{x:,.2f}")
ax.yaxis.set_major_formatter(tf) 
plt.title("Total gasto por UF")

plt.show()
```
